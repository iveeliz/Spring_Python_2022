## Exercise 1  
**<font color='forestgreen'>1.</font>** <font color='forestgreen'>Создайте график $x^2⋅sin(1/x^2)+x$ на интервале $[−1,1]$, используя 250 точек. Не забудьте пометить оси!</font>


```python
x = np.linspace(-1, 1, 250)
y = x**2 * np.sin(1/x**2) + x
```

Когда данные для графика подготовлены, нам нужно создать экземпляры Matplotlib *Figure* as *fig* и *Axes* as *ax*, затем использовать метод ax.plot для построения данных и установить основные свойства графика.

```python
fig, ax = plt.subplots(figsize=(10, 7))

ax.plot(x, y, color = 'violet', label = 'y = $x^2\sin(1/x^2) + x$')
ax.legend(fontsize = 18)

ax.set_xlabel(r'$x$', fontsize = 18, color = 'mediumvioletred')
ax.set_ylabel(r'$y$', fontsize = 18, color = 'mediumvioletred')
ax.set_title('\nГрафик функции y(x), $x \in [-1; 1]$\n', color = 'rebeccapurple', fontsize = 18)
ax.grid(color = 'gainsboro')

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['left'].set_color('none')
ax.spines['bottom'].set_color('none')
```

**<font color='forestgreen'>2.</font>** <font color='forestgreen'>Создайте semilogy график (с логарифмической шкалой для оси y и линейной шкалой для оси x) относительной разницы  функций $1/(1+x^2)$ и $1/x^2$ на интервале $[5,25]$. **<font color='forestgreen'>1.</font>** <font color='forestgreen'>Построить график функции $sin(x)/x$ на отрезке $[-6\pi;6\pi]$ и добавить тики по оси $x$ с шагом $\pi/2$ от $-6\pi$ до $6\pi$, причём обозначения тиков должны содержать греческую букву $\pi$. Добавить сетку на график.</font>

```python
x = np.linspace(-6 * np.pi, 6 * np.pi, 250)
y = np.sin(x)/x
```

```python
fig, ax = plt.subplots(figsize=(17, 5))

A = [r'$-6\pi$', r'$-\frac {11\pi}{2}$', r'$-5\pi$', r'$-\frac {9\pi}{2}$', r'$-4\pi$', r'$-\frac {7\pi}{2}$', 
     r'$-3\pi$', r'$-\frac {5\pi}{2}$', r'$-2\pi$', r'$-\frac {3\pi}{2}$', r'$-\pi$', r'$-\frac {\pi}{2}$',
     r'$0$', r'$\frac {\pi}{2}$', r'$\pi$', r'$\frac {3\pi}{2}$', r'$2\pi$', r'$\frac {5\pi}{2}$', r'$3\pi$',
     r'$\frac {7\pi}{2}$', r'$4\pi$', r'$\frac {9\pi}{2}$', r'$5\pi$', r'$\frac {11\pi}{2}$', r'$6\pi$']

ax.set_xticks(np.linspace(-6 * np.pi, 6 * np.pi, 25))
ax.set_xticklabels(A, fontsize = 14)

ax.plot(x, y, color = 'violet', label = 'y = $\sin(x)/x$')
ax.legend(fontsize = 18)

ax.set_xlabel(r'$x$', fontsize = 18, color = 'mediumvioletred')
ax.set_ylabel(r'$y$', fontsize = 18, color = 'mediumvioletred')
ax.set_title('\nГрафик функции y(x) = $\sin(x)/x$, $x \in [-6\pi; 6\pi]$\n', color = 'rebeccapurple', fontsize = 18)
ax.grid(color = 'gainsboro')

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['left'].set_color('none')
ax.spines['bottom'].set_color('none')
```

```python
fig, ax = plt.subplots(figsize=(17, 5))

ax.set_xticks(np.linspace(-6 * np.pi, 6 * np.pi, 25))

A = []
a = -6

while a <= 6:
    
    if a % 1 == 0 and a != -1 and a != 0 and a != 1:
        A.append(str(int(a)) + '$\pi$')
        
    if a % 1 == 0.5 and a != -0.5 and a != 0.5:
        a = 2 * a
        A.append(str(int(a)) + '$\pi/2$')
        a = a/2
    
    if a == -1:
        A.append('$-\pi$')
    
    if a == -0.5:
        A.append('$-\pi/2$')
        
    if a == 0:
        A.append('$0$')
        
    if a == 0.5:
        A.append('$\pi/2$')
    
    if a == 1:
        A.append('$\pi$')
    
    a += 0.5
    
ax.set_xticklabels(A)

ax.plot(x, y, color = 'violet', label = 'y = $\sin(x)/x$')
ax.legend(fontsize = 18)

ax.set_xlabel(r'$x$', fontsize = 18, color = 'mediumvioletred')
ax.set_ylabel(r'$y$', fontsize = 18, color = 'mediumvioletred')
ax.set_title('\nГрафик функции y(x) = $\sin(x)/x$, $x \in [-6\pi; 6\pi]$\n', color = 'rebeccapurple', fontsize = 18)
ax.grid(color = 'gainsboro')

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['left'].set_color('none')
ax.spines['bottom'].set_color('none')
```

## Exercise 3  
**<font color='forestgreen'>1.</font>** <font color='forestgreen'>Создайте цветовой график $sin(x)sin(y)$ на интервале $[−π,π]×[−π,π]$.</font>

```python
x = np.linspace(-np.pi, np.pi, 250)
y = np.linspace(-np.pi, np.pi, 250)

X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.sin(Y)
```

```python
fig, ax = plt.subplots(figsize=(10, 9))

norm = mpl.colors.Normalize(-abs(Z).max(), abs(Z).max())
p = ax.pcolor(X, Y, Z, norm = norm,  cmap = 'Pastel1', shading='auto')

cb = fig.colorbar(p, ax=ax)
cb.set_label(r'$z$', fontsize = 18, color = 'mediumvioletred')
cb.set_ticks([-1, -0.75, -0.5, -0.25, 0, 0.25, 0.5, 0.75, 1])

ax.set_xlabel(r'$x$', fontsize = 18, color = 'mediumvioletred')
ax.set_ylabel(r'$y$', fontsize = 18, color = 'mediumvioletred')

ax.xaxis.set_major_locator(mpl.ticker.MaxNLocator(7))
ax.yaxis.set_major_locator(mpl.ticker.MaxNLocator(7))

ax.set_title('\nz(x, y) = $sin(x)sin(y)$, $x, y \in [−π,π]×[−π,π]$\n', color = 'rebeccapurple', fontsize = 18)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 9), subplot_kw = {'projection': '3d'})

norm = mpl.colors.Normalize(-abs(Z).max(), abs(Z).max())
p = ax.plot_surface(X, Y, Z, norm = norm,  cmap = 'cool', rstride = 1, cstride = 1)

ax.set_xlabel(r'$x$', fontsize = 18, color = 'mediumvioletred')
ax.set_ylabel(r'$y$', fontsize = 18, color = 'mediumvioletred')
ax.set_zlabel(r'$z$', fontsize = 18, color = 'mediumvioletred')

ax.xaxis.set_major_locator(mpl.ticker.MaxNLocator(7))
ax.yaxis.set_major_locator(mpl.ticker.MaxNLocator(7))

ax.set_title('\nz(x, y) = $sin(x)sin(y)$, $x, y \in [−π,π]×[−π,π]$\n', color = 'rebeccapurple', fontsize = 18)
```

**<font color='forestgreen'>2.</font>** <font color='forestgreen'>Создайте цветовой график $e^{-x^2-y^2}$ на интервале $[−3,3]×[−3,3]$.</font>

```python
x = np.linspace(-3, 3, 250)
y = np.linspace(-3, 3, 250)

X, Y = np.meshgrid(x, y)
Z = np.exp(- X**2 - Y**2)
```

```python
fig, ax = plt.subplots(figsize=(10, 9))

norm = mpl.colors.Normalize(-abs(Z).max(), abs(Z).max())
p = ax.pcolor(X, Y, Z, norm = norm,  cmap = 'cool', shading='auto')

cb = fig.colorbar(p, ax=ax)
cb.set_label(r'$z$', fontsize = 18, color = 'mediumvioletred')
cb.set_ticks([-1, -0.75, -0.5, -0.25, 0, 0.25, 0.5, 0.75, 1])

ax.set_xlabel(r'$x$', fontsize = 18, color = 'mediumvioletred')
ax.set_ylabel(r'$y$', fontsize = 18, color = 'mediumvioletred')

ax.xaxis.set_major_locator(mpl.ticker.MaxNLocator(7))
ax.yaxis.set_major_locator(mpl.ticker.MaxNLocator(7))

ax.set_title('\nz(x, y) = $e^{- x^2 - y^2}$, $x$, $y \in [−3,3]×[−3,3]$\n', color = 'rebeccapurple', fontsize = 18)
```
