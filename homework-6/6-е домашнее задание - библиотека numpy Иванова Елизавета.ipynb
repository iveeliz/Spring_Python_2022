6.  Создайте функцию, которая создает $n×n$ матрицу с $(i,j)$-элементом, равным $i+j$.

```python
def func1(n):
    
    A = np.zeros((n, n))
    
    for i in range(n):
        for j in range(n):
            
            A[i][j] = i + j
    
    return A

func1(10)
```

```python
def func2(n):
    
    B = np.arange(0, n**2).reshape((n, n))
    
    B = B // n + B % n
    
    return B

func2(10)
```

## Аналитическое решение:

$\int\limits_0^2 {\frac{{2x - 1}}{{2x + 1}}dx}$ = $ \frac{1}{2} \int {\frac{{t - 2}}{{t}}dt}$ = $ \frac{1}{2} \int {1 - {\frac{{2}}{{t}}}dt}$ = $ \frac{1}{2}({t} - {2 \ln {\mid {t} \mid}})$ = $ (\frac{1}{2}({2x + 1} - {2 \ln {\mid {2x + 1} \mid}})) \mid_0^2 $ = $ ({x} + \frac{1}{2} - \ln {\mid {2x + 1} \mid}) \mid_0^2 $ = $ 2 - \ln {5} \approx  0.3905620875658996 $

```python
# Подынтегральное выражение (функция)

def integrand(x):
    
    return (2 * x - 1) / (2 * x + 1)
```

```python
# Метод прямоугольников

def rectangle(function, a, b, step_int):
    
    dx = (b - a) / step_int
    A = np.linspace(a, b - step_int, int(dx)) # возвращает равномерно распределенные числа из указанного интервала
    f = integrand(A) * step_int
    
    return np.sum(f)
```

```python
# Метод трапеций

def trapezium(function, a, b, step_int):
    
    dx = (b - a) / step_int
    A = np.linspace(a, b, int(dx) + 1)
    f = integrand(A)
    
    return step_int / 2 * (f[0] + 2 * np.sum(f[1 : len(A) - 1]) + f[-1])
```

```python
# Правило Симпсона

def Simpson(function, a, b, step_int):
    
    dx = (b - a) / step_int
    A = np.linspace(a, b, int(dx) + 1)
    N = len(A) - 1
    f = integrand(A)
    
    return step_int / 3 * (f[0] + 2 * np.sum(f[2 : N : 2])  + 4 * np.sum(f[1 : N : 2]) + f[N])
```

```python
print('Метод прямоугольников, шаг интегрирования = 0.1: ', rectangle(integrand, 0, 2, 0.1))
print('Метод трапеций, шаг интегрирования = 0.1:        ', trapezium(integrand, 0, 2, 0.1))
print('Метод Симпсона, шаг интегрирования = 0.1:        ', Simpson(integrand, 0, 2, 0.1))

print('\nМетод прямоугольников, шаг интегрирования = 0.01: ', rectangle(integrand, 0, 2, 0.01))
print('Метод трапеций, шаг интегрирования = 0.01:        ', trapezium(integrand, 0, 2, 0.01))
print('Метод Симпсона, шаг интегрирования = 0.01:        ', Simpson(integrand, 0, 2, 0.01))

print('\nМетод прямоугольников, шаг интегрирования = 0.001: ', rectangle(integrand, 0, 2, 0.001))
print('Метод трапеций, шаг интегрирования = 0.001:        ', trapezium(integrand, 0, 2, 0.001))
print('Метод Симпсона, шаг интегрирования = 0.001:        ', Simpson(integrand, 0, 2, 0.001))

print('\nМетод прямоугольников, шаг интегрирования = 0.0001: ', rectangle(integrand, 0, 2, 0.0001))
print('Метод трапеций, шаг интегрирования = 0.0001:        ', trapezium(integrand, 0, 2, 0.0001))
print('Метод Симпсона, шаг интегрирования = 0.0001:        ', Simpson(integrand, 0, 2, 0.0001))

print('\nМетод прямоугольников, шаг интегрирования = 0.00001: ', rectangle(integrand, 0, 2, 0.00001))
print('Метод трапеций, шаг интегрирования = 0.00001:        ', trapezium(integrand, 0, 2, 0.00001))
print('Метод Симпсона, шаг интегрирования = 0.00001:        ', Simpson(integrand, 0, 2, 0.00001))

print('\nМетод прямоугольников, шаг интегрирования = 0.000001: ', rectangle(integrand, 0, 2, 0.000001))
print('Метод трапеций, шаг интегрирования = 0.000001:        ', trapezium(integrand, 0, 2, 0.000001))
print('Метод Симпсона, шаг интегрирования = 0.000001:        ', Simpson(integrand, 0, 2, 0.000001))

print('\nМетод прямоугольников, шаг интегрирования = 0.0000001: ', rectangle(integrand, 0, 2, 0.0000001))
print('Метод трапеций, шаг интегрирования = 0.0000001:        ', trapezium(integrand, 0, 2, 0.0000001))
print('Метод Симпсона, шаг интегрирования = 0.0000001:        ', Simpson(integrand, 0, 2, 0.0000001))
```
